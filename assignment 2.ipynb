```python
 1. # Program to calculate area of square and circle 

# Square formula : side * side
side = float(input("Enter the side of the square: "))
area_square = side * side
print("Area of Square =", area_square)
```

```python
# Circle formula: 3.14 * r * r
radius = float(input("Enter the radius of the circle: "))
area_circle = 3.14 * radius * radius
print("Area of Circle =", area_circle)
```

```python
2. # Write a program that asks the user for the side of a square and prints its perimeter.
# perimeter : 4 * side
side=float(input("Enter the side of the square: "))
perimeter = 4*side
print("Square of the perimeter is=", perimeter)
```

```python
3. #Take the base and height of a triangle as input and print its area.
# formula:1/2 * base * height
Base=10.2
height=14.5
area_triangle=1/2 * Base * height
print("Area of triangle is=", area_triangle)
```

```python
4. #Write a program that asks the user for the radius of a circle and prints its circumference. (Use 3.14 for π).
# circumference formula : 2 * 3.14 * r

radius=float(input("Enter radius of a circle :"))
circumference=2*3.14*radius
print("Circumference of a circle is=", circumference)
```

```python
5. #Take Principal (P), Rate (R), and Time (T) as input from the user and print the Simple Interest.
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time: "))
simple_interest=principal*rate*time/100
print("Simple Interest is=",simple_interest)
```
